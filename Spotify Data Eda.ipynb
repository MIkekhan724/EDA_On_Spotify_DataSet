Chack Correlation Between Variable in the Dataset

```python
corr_df=df_track.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
#sns.set(rc={'figure.figsize':(20,7)})
plt.figure(figsize=(14,6))
heat_map=sns.heatmap(corr_df,annot=True,fmt='.1g',vmin=-1,vmax=1,center=0,cmap='inferno',linewidths=1,linecolor='Black')
heat_map.set_title("Correlation Heatmap Between Variables")
heat_map.set_xticklabels(heat_map.get_xticklabels(),rotation=90)
plt.show()
```

annot=True -->annot to add numbers on the box

fmt='.1g' -->string formating for annot

values to ancor the colur map vmin=-1 vmax=1

cmap is for colur map it's heatmap design

```python
# sampling the data to create regression plot .4% of data
sample_df=df_track.sample(int(0.004*len(df_track)))
```

```python
print(len(sample_df))
```

lets Plot regression line between loudness and energy

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,x='energy',y='loudness',color='c').set(title='Loudness vs Energy Correlation')
plt.show()
```

 It's shows that there is high correlation it's mean if loudness increase the energy alse increases vice versa

```python
# popularity vs acousticness
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,x='popularity',y='acousticness',color='b').set(title='popularity vs acousticness Correlation')
plt.show()
```

It's shows there is negative correlation between popularity and acousticness

it mean if  acousticness increses then popularity decreases 

```python
#extracting Year from released date
df_track['dates']=df_track.index.get_level_values('release_date')
df_track.dates=pd.to_datetime(df_track.dates)
```

```python
years=df_track.dates.dt.year
years
```

```python
#Numbers Of Songs Per Year Using Distribution plot
sns.set(rc={'figure.figsize':(20,7)})
sns.distplot(years, color='b').set(title="Numbers of songs Per Year")
plt.show()
```

```python
#number of songs per year using Histogram
sns.set(rc={'figure.figsize':(20,7)})
sns.histplot(data=years, bins='auto',  color='g',  legend=True)
plt.show()
```

```python
# Importing Our Second DataSet
df_genre=pd.read_csv('SpotifyFeatures.csv')
df_genre.head()
```

```python
# Duration of the Songs for different Genre
sns.set(rc={'figure.figsize':(20,10)})
plt.title('Duration Of the Songs in different Genre')
sns.color_palette('rocket',as_cmap=True)
sns.barplot(y='genre',x='duration_ms',data=df_genre)
plt.xlabel('Duration in Millisecond')
plt.ylabel('Genre')
plt.show()
```

```python
# Top 5 Genre By Popularity
sns.set_style(style='darkgrid')
sns.set(rc={'figure.figsize':(20,10)})
famous=df_genre.sort_values("popularity",ascending=False).head(10)
sns.barplot(y='genre',x='popularity',data=famous).set(title='Top 5 Genre by Popularity')
plt.show()
```

These are our Top 5 genre by Popularity
